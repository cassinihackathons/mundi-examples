# Mundi Easy Search Sentinel 2 Image

This Jupyter Notebook "Mundi Easy Search Sentinel 2 Image" offers you the possibility to find and save an image from the Mundi Catalogue in the easiest way! You can choose the geographical zone, the WMS layers and the time interval or date you want!

This notebook is based on WMS to find an image given an area of interest:
* If you specify a time interval, it will choose the image with least clouds in this interval
* If you specify a date, it will choose an image based on two criteria: least clouds and closest to the date

The downloaded images will be stored in `work/sentinel2_images`

```python
from imglib import find_image, sentinel_images_display, sentinel_images_delete, setup
setup()
```

## Your image search parameters :

### 1/ Bounding Box :

A bounding box is a geographical area defined by two longitudes and two latitudes. To select your bounding box, you can either use the interface, and type the city you want to see, or directly define your own bounding box.

Warning, the bounding box cannot be too large ! You cannot for instance specify a country such as France as a bounding box. The size of a city is more suitable. 

To get a bounding box around the place you want, you can go on www.bboxfinder.com  
- Select EPSG:4326 as the system coordinates, to get the right bbox
- Go to the place you want to get the images
- Copy the coordinates next to the Map box

### 2/ Image Parameters :

You have various parameters to create your image :
- Collections (satellites) : 
    - Sentinel2-L1C
    - Sentinel2-L2A
- Layer : The layer you want to see in your image
- Starting Date & final date : The timedelta where we will search for the image or the date where we will search for the nearest best image
- Image title : The title you want to give to your image 
- File Name : The name you want to give to your image (mandatory field)
- Height & Width : Height and Width (in pixels) of the image in output

When you click on download, the image will be directly put on your `/work/sentinel2_images/` folder
