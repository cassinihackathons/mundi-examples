# Mundi WMS City Polygon

```python
from utils import city_polygon_bbox
from utils import display_wms
from mundilib import MundiCatalogue
```

### Selection of the area 

```python
location = 'Toulouse'  # examples : Carmes, Marseilles, Monaco, Switzerland
polygon, bbox, place_name = city_polygon_bbox(location)
print(f'Name of the place = {place_name}')
print(f'Bounding box (Bbox) = {bbox}')
```

### Display the polygon and the satellite image of the selected area

```python
# wms_layers can take only two values ('1' for Natural color or 'All' for all layers)
c = MundiCatalogue()
wms = c.get_collection('Sentinel2').mundi_wms('L1C')
display_wms(polygon,bbox,wms,wms_layers='1',time ='2021-12-27/2022-01-10')
```
