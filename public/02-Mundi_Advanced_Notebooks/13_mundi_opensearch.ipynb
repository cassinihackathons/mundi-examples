# Mundi Opensearch

```python
from mundilib import MundiCatalogue, MundiCollection, find_entries
```

### OpenSearch 'Get' request example
#### The query we'll use

```python
#https://sentinel3.browse.catalog.mundiwebservices.com/opensearch?q=(onlineStatus:(ONLINE OR STAGING) 
#AND sensingStartDate:[2018-05-20T00:00:00Z TO 2018-05-20T23:59:59Z])&processingLevel=L2_

from mundilib import opensearch_query

q = "q=(sensingStartDate:[2022-05-20T00:00:00Z TO 2022-05-20T00:00:05Z])&processingLevel=L2_"
```

#### Launch an OpenSearch request on Sentinel3 MundiCollection

```python
c = MundiCatalogue()
s3 = c.get_collection("Sentinel3")
```

```python
response_wrapper_list = opensearch_query(collection=s3, query=q)
```

```python
entries = find_entries(response_wrapper_list)
print (len(entries))
```

#### Getting specific 'entry' field content

```python
for entry in entries[:3]:
    node = entry.find("eo:processingDate", namespaces=entry.nsmap)
    print(node.text)
    node = entry.find("eo:processingLevel", namespaces=entry.nsmap)
    print(node.text)
```

#### Display all 'opensearch' fields content from an entry

```python
obj = entries[:1][0]
d = {}
for childNodes in obj:
    attr = str(childNodes.prefix) + ":" + childNodes.tag[childNodes.tag.index('}')+1:]
    d[attr] = childNodes.text

# list some attribute/value of the entry (panda table)
import pandas as pd
df = pd.DataFrame(d, index=['attribute value'])
pd.set_option('max_colwidth', 200)
df.T
```

#### Only get number of opensearch request products

```python
from mundilib import opensearch_nb_results

nb = opensearch_nb_results(s3, query=q)
print (nb)
```
